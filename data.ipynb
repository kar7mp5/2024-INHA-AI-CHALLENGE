```python
import pandas as pd
import numpy as np
import csv
import os
```

### **Load the data**

```python
train_path: str = os.path.join(os.getcwd(), 'contents', 'train.csv')

train_data = pd.read_csv(train_path)
```

```python
train_data.head(5)
```

```python
train_data.drop_duplicates(['context'])
```

```python
print(train_data['context'][1])
print(train_data['context'][2])
```

### **Merge the data**

```python
print(train_data['context'][1] + train_data['context'][2])
```

```python
train_data.shape
```

```python
# Assuming train_data is already defined and is a DataFrame
df_cleaning = pd.DataFrame(columns=['id', 'context', 'question', 'answer'])

# Check the number that is odd.
data_length: int = [train_data.shape[0]  if train_data.shape[0]//2 == 0 else train_data.shape[0] -1][0]
print(f"Data Length: {data_length}")
for i in range(1, data_length):    
    # Creating a new Series with the combined context
    new_row = pd.Series({'context': train_data['context'][i] + train_data['context'][i+1], 'id': train_data['id'][i], 'question': train_data['question'][i], 'answer': train_data['answer'][i]})

    # Appending the new row to the DataFrame
    df_cleaning = pd.concat([df_cleaning, new_row.to_frame().T], ignore_index=True)

df_cleaning.head(2)
```

```python
print(df_cleaning['context'][0])
```

### **Save the model**

```python
df_cleaning.to_csv(os.path.join('contents', 'crossover.csv'))
```

```python
batch = df_cleaning.shape[0]//10

for i in range(1, df_cleaning.shape[0]//batch + 1):
    df_clone = df_cleaning.iloc[(i-1)*batch: i*batch, :]
    print(df_clone.shape)
    df_clone.to_csv(os.path.join('contents', f'{i}_crossover.csv'))   
```
